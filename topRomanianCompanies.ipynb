```python
import pandas as pd

companies = pd.read_html("https://en.wikipedia.org/wiki/List_of_companies_of_Romania")[0]
companies
```

```python
def most_popular_headquarters(companies: pd.DataFrame) -> str:
    """Returns the city in that the most companies have their headquarters. 

    Args:
        companies (pd.DataFrame): the dataset

    Returns:
        str: the name of the city
    """
    return companies.groupby(by=["Headquarters"]).count().sort_values(by=["Name", "Headquarters"], ascending=[False, True]).reset_index().loc[0, "Headquarters"]

most_popular_headquarters(companies=companies)
```

```python
def most_popular_industries(companies: pd.DataFrame) -> str:
    """Returns the industry in that the most companies operate. 

    Args:
        companies (pd.DataFrame): the dataset

    Returns:
        str: the name of the industry
    """
    return companies.groupby(by=["Industry"]).count().sort_values(by=["Name", "Industry"], ascending=[False, True]).reset_index().loc[0, "Industry"]

most_popular_industries(companies=companies)
```
